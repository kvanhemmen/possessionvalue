```python
def scrape_club(url, path_to_chromedriver=""):
    # imports necessary libraries
    ## some must be downloaded
    import urllib.request as urllib
    import time
    import random
    from selenium import webdriver
    import json
    from bs4 import BeautifulSoup
    import bs4 as bs
    
    # instantiate the webdriver
    driver = webdriver.Chrome()
    # get the fixture list
    driver.get(url)           
    html = driver.page_source
    soup = bs.BeautifulSoup(html)
    driver.close()
    # get all of the fixtures as an iterable object
    links = soup.findAll('a', class_="horiz-match-link result-1")
    i=0
    for link in links:
        # isolate the link
        fpage = link.get("href").strip()
        name = fpage[34:]
        # print the fixtures so you know what you're getting
        print(name)
        # this requires tweaking if you aren't scraping Eredivisie data! It ensures that only league matches are retrieved
        if name[0:10] == 'Eredivisie':
            i+=1
            url = "https://www.whoscored.com" + fpage
            # instantiate webdriver
            driver = webdriver.Safari()
            driver.get(url) 
            # access the page
            html = driver.page_source
            soup = BeautifulSoup(html)
            driver.close()
            scripts = soup.find_all("script")
            # find the event data
            for tag in scripts:
                if "matchCentreData" in tag.get_text():
                    data = tag.get_text()
            data = data[32:-4975]
            # write it to a csv on your desktop
            filename= 'Desktop/match' + str(i) + '.txt'
            text_file = open(filename, "w")
            n = text_file.write(data)
            text_file.close()
```

```python
scrape_club("https://www.whoscored.com/Teams/256/Fixtures/Netherlands-Feyenoord")
```
